```python
# BONUS ASSIGNMENT - PART 1: Implement a 1D CNN (20 points)
# Same as Assignment 06 â€” included for continuity

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dense, Flatten
from tensorflow.keras.utils import to_categorical

print("ğŸš€ BONUS ASSIGNMENT - PART 1: 1D CNN Implementation")
print("=" * 70)
print("ğŸ“‹ Assignment Overview:")
print("   â€¢ Dataset: Iris dataset (same as Assignment 06)")
print("   â€¢ Task: 1D CNN for classification")
print("   â€¢ Points: 20")
print("   â€¢ Purpose: Continuity with previous assignment")
print("=" * 70)

# ============================================================================
# STEP 1: Dataset Loading
# ============================================================================

def load_iris_dataset():
    """
    Load the Iris dataset

    Returns:
        X: Features array
        y: Target array
        feature_names: Names of features
        target_names: Names of target classes
    """

    print("\nğŸ“Š STEP 1: Loading Iris Dataset")
    print("-" * 40)

    # Load the dataset
    iris = load_iris()
    X = iris.data
    y = iris.target
    feature_names = iris.feature_names
    target_names = iris.target_names

    print(f"âœ… Dataset loaded successfully")
    print(f"   â€¢ Samples: {X.shape[0]}")
    print(f"   â€¢ Features: {X.shape[1]}")
    print(f"   â€¢ Classes: {len(target_names)}")
    print(f"   â€¢ Feature names: {feature_names}")
    print(f"   â€¢ Target names: {target_names}")

    return X, y, feature_names, target_names

# ============================================================================
# STEP 2: Preprocessing
# ============================================================================

def preprocess_iris_data(X, y, test_size=0.2, random_state=42):
    """
    Preprocess the Iris dataset according to bonus requirements

    Args:
        X: Features array
        y: Target array
        test_size: Test set size (default: 0.2)
        random_state: Random seed for reproducibility

    Returns:
        Preprocessed training and testing data
    """

    print("\nğŸ”§ STEP 2: Preprocessing Data")
    print("-" * 40)

    # Split the data first
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )

    print(f"âœ… Data split completed:")
    print(f"   â€¢ Training samples: {X_train.shape[0]}")
    print(f"   â€¢ Testing samples: {X_test.shape[0]}")

    # ========================================================================
    # 2a. Normalize feature values
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 2a: Normalize feature values")
    print("-" * 30)

    scaler = StandardScaler()
    X_train_normalized = scaler.fit_transform(X_train)
    X_test_normalized = scaler.transform(X_test)

    print(f"âœ… Feature normalization completed")
    print(f"   â€¢ Method: StandardScaler (mean=0, std=1)")
    print(f"   â€¢ Training mean: {np.mean(X_train_normalized, axis=0)}")
    print(f"   â€¢ Training std: {np.std(X_train_normalized, axis=0)}")

    # ========================================================================
    # 2b. One-hot encode target labels
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 2b: One-hot encode target labels")
    print("-" * 30)

    y_train_onehot = to_categorical(y_train, num_classes=3)
    y_test_onehot = to_categorical(y_test, num_classes=3)

    print(f"âœ… One-hot encoding completed")
    print(f"   â€¢ Original shape: {y_train.shape}")
    print(f"   â€¢ One-hot shape: {y_train_onehot.shape}")
    print(f"   â€¢ Classes: 3 (setosa, versicolor, virginica)")

    # ========================================================================
    # 2c. Reshape input for CNN: (samples, features, 1)
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 2c: Reshape input for CNN")
    print("-" * 30)

    X_train_cnn = X_train_normalized.reshape(X_train_normalized.shape[0], X_train_normalized.shape[1], 1)
    X_test_cnn = X_test_normalized.reshape(X_test_normalized.shape[0], X_test_normalized.shape[1], 1)

    print(f"âœ… CNN reshaping completed")
    print(f"   â€¢ Original shape: {X_train_normalized.shape}")
    print(f"   â€¢ CNN shape: {X_train_cnn.shape}")
    print(f"   â€¢ Format: (samples, features, 1)")

    return {
        'X_train': X_train_cnn,
        'X_test': X_test_cnn,
        'y_train': y_train_onehot,
        'y_test': y_test_onehot,
        'y_train_original': y_train,
        'y_test_original': y_test,
        'scaler': scaler
    }

# ============================================================================
# STEP 3: Model Architecture
# ============================================================================

def build_1d_cnn_model(input_shape):
    """
    Build 1D CNN model according to bonus requirements

    Args:
        input_shape: Shape of input data (features, channels)

    Returns:
        Compiled CNN model
    """

    print("\nğŸ—ï¸ STEP 3: Model Architecture")
    print("-" * 40)

    model = Sequential()

    # ========================================================================
    # 3a. 1D Convolutional Layer(s)
    # ========================================================================
    print(f"ğŸ“‹ INSTRUCTION 3a: 1D Convolutional Layer(s)")

    model.add(Conv1D(
        filters=32,
        kernel_size=2,
        activation='relu',
        input_shape=input_shape,
        name='conv1d_layer1'
    ))

    # Additional Conv1D layer for better feature extraction
    model.add(Conv1D(
        filters=64,
        kernel_size=2,
        activation='relu',
        name='conv1d_layer2'
    ))

    print(f"âœ… 1D Convolutional layers added:")
    print(f"   â€¢ Layer 1: 32 filters, kernel_size=2, ReLU")
    print(f"   â€¢ Layer 2: 64 filters, kernel_size=2, ReLU")

    # ========================================================================
    # 3b. MaxPooling
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 3b: MaxPooling")

    model.add(MaxPooling1D(
        pool_size=2,
        name='maxpool1d_layer'
    ))

    print(f"âœ… MaxPooling layer added:")
    print(f"   â€¢ Pool size: 2")
    print(f"   â€¢ Purpose: Dimensionality reduction")

    # Flatten for dense layers
    model.add(Flatten(name='flatten_layer'))

    # ========================================================================
    # 3c. Dense output layer
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 3c: Dense output layer")

    # Hidden dense layer
    model.add(Dense(
        64,
        activation='relu',
        name='dense_hidden'
    ))

    # Output layer
    model.add(Dense(
        3,
        activation='softmax',
        name='dense_output'
    ))

    print(f"âœ… Dense layers added:")
    print(f"   â€¢ Hidden layer: 64 neurons, ReLU")
    print(f"   â€¢ Output layer: 3 neurons, Softmax (for 3 classes)")

    return model

def compile_and_display_model(model):
    """
    Compile model and display architecture summary

    Args:
        model: Built CNN model

    Returns:
        Compiled model
    """

    # Compile the model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',  # Using one-hot encoded labels
        metrics=['accuracy']
    )

    print(f"\nğŸ“Š Model Compilation:")
    print(f"   â€¢ Optimizer: Adam")
    print(f"   â€¢ Loss: categorical_crossentropy (for one-hot labels)")
    print(f"   â€¢ Metrics: accuracy")

    print(f"\nğŸ—ï¸ Model Architecture Summary:")
    print("-" * 50)
    model.summary()

    return model

# ============================================================================
# STEP 4: Train and Evaluate
# ============================================================================

def train_cnn_model(model, data, epochs=20, batch_size=16, validation_split=0.2):
    """
    Train the CNN model

    Args:
        model: Compiled CNN model
        data: Preprocessed data dictionary
        epochs: Number of training epochs
        batch_size: Training batch size
        validation_split: Validation split ratio

    Returns:
        Training history
    """

    print(f"\nğŸš€ STEP 4a: Training the Model")
    print("-" * 40)

    print(f"ğŸ“‹ Training Configuration:")
    print(f"   â€¢ Epochs: {epochs}")
    print(f"   â€¢ Batch size: {batch_size}")
    print(f"   â€¢ Validation split: {validation_split}")

    # Train the model
    history = model.fit(
        data['X_train'], data['y_train'],
        epochs=epochs,
        batch_size=batch_size,
        validation_split=validation_split,
        verbose=1
    )

    print(f"\nâœ… Training completed successfully!")

    return history

def evaluate_cnn_model(model, data, target_names):
    """
    Evaluate the CNN model

    Args:
        model: Trained CNN model
        data: Preprocessed data dictionary
        target_names: Names of target classes

    Returns:
        Evaluation results
    """

    print(f"\nğŸ“Š STEP 4b: Evaluating the Model")
    print("-" * 40)

    # Evaluate on test set
    test_loss, test_accuracy = model.evaluate(data['X_test'], data['y_test'], verbose=0)

    print(f"âœ… Test Evaluation Results:")
    print(f"   â€¢ Test Loss: {test_loss:.4f}")
    print(f"   â€¢ Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

    # Get predictions for detailed analysis
    y_pred_proba = model.predict(data['X_test'], verbose=0)
    y_pred = np.argmax(y_pred_proba, axis=1)
    y_true = data['y_test_original']

    # Classification report
    print(f"\nğŸ“‹ Detailed Classification Report:")
    print("-" * 40)
    print(classification_report(y_true, y_pred, target_names=target_names))

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    return {
        'test_loss': test_loss,
        'test_accuracy': test_accuracy,
        'y_pred': y_pred,
        'y_true': y_true,
        'confusion_matrix': cm
    }

def analyze_training_performance(history, evaluation_results):
    """
    Analyze training performance including accuracy, loss curves, convergence and validation

    Args:
        history: Training history object
        evaluation_results: Model evaluation results
    """

    print(f"\nğŸ“‹ INSTRUCTION 4: Train and Evaluate - Detailed Analysis")
    print("=" * 60)

    # Extract training metrics
    train_acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = len(train_acc)

    # ========================================================================
    # 4a. Accuracy Analysis
    # ========================================================================
    print(f"\n[ACCURACY] Accuracy Curve Analysis:")
    print("-" * 40)

    initial_train_acc = train_acc[0]
    final_train_acc = train_acc[-1]
    best_val_acc = max(val_acc)
    best_val_acc_epoch = np.argmax(val_acc) + 1
    final_val_acc = val_acc[-1]
    test_acc = evaluation_results['test_accuracy']

    print(f"â€¢ Initial training accuracy: {initial_train_acc:.4f} ({initial_train_acc*100:.2f}%)")
    print(f"â€¢ Final training accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")
    print(f"â€¢ Best validation accuracy: {best_val_acc:.4f} at epoch {best_val_acc_epoch}")
    print(f"â€¢ Final validation accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)")
    print(f"â€¢ Test accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)")

    accuracy_improvement = final_train_acc - initial_train_acc
    print(f"â€¢ Training accuracy improvement: {accuracy_improvement:.4f} ({accuracy_improvement*100:.2f}%)")

    # ========================================================================
    # 4b. Loss Curve Analysis
    # ========================================================================
    print(f"\n[LOSS] Loss Curve Analysis:")
    print("-" * 40)

    initial_train_loss = train_loss[0]
    final_train_loss = train_loss[-1]
    min_val_loss = min(val_loss)
    min_val_loss_epoch = np.argmin(val_loss) + 1
    final_val_loss = val_loss[-1]
    test_loss = evaluation_results['test_loss']

    print(f"â€¢ Initial training loss: {initial_train_loss:.4f}")
    print(f"â€¢ Final training loss: {final_train_loss:.4f}")
    print(f"â€¢ Minimum validation loss: {min_val_loss:.4f} at epoch {min_val_loss_epoch}")
    print(f"â€¢ Final validation loss: {final_val_loss:.4f}")
    print(f"â€¢ Test loss: {test_loss:.4f}")

    loss_reduction = initial_train_loss - final_train_loss
    loss_reduction_pct = (loss_reduction / initial_train_loss) * 100
    print(f"â€¢ Training loss reduction: {loss_reduction:.4f} ({loss_reduction_pct:.1f}%)")

    # ========================================================================
    # 4c. Convergence Analysis
    # ========================================================================
    print(f"\n[CONVERGENCE] Convergence Analysis:")
    print("-" * 40)

    # Check if training is converging (last 3 epochs stability)
    if epochs >= 3:
        last_3_train_acc = train_acc[-3:]
        last_3_val_acc = val_acc[-3:]
        last_3_train_loss = train_loss[-3:]
        last_3_val_loss = val_loss[-3:]

        acc_stability = np.std(last_3_val_acc)
        loss_stability = np.std(last_3_val_loss)

        print(f"â€¢ Validation accuracy stability (last 3 epochs): {acc_stability:.6f}")
        print(f"â€¢ Validation loss stability (last 3 epochs): {loss_stability:.6f}")

        if acc_stability < 0.01 and loss_stability < 0.01:
            convergence_status = "CONVERGED - Model has stabilized"
        elif acc_stability < 0.05 and loss_stability < 0.05:
            convergence_status = "CONVERGING - Model is stabilizing"
        else:
            convergence_status = "TRAINING - Model still improving"

        print(f"â€¢ Convergence status: {convergence_status}")

        # Early stopping analysis
        if best_val_acc_epoch < epochs * 0.8:
            print(f"â€¢ Early stopping recommendation: Could stop at epoch {best_val_acc_epoch}")
        else:
            print(f"â€¢ Training duration: Appropriate length")

    # ========================================================================
    # 4d. Validation Performance Analysis
    # ========================================================================
    print(f"\n[VALIDATION] Validation Performance Analysis:")
    print("-" * 40)

    # Overfitting analysis
    train_val_gap = abs(final_train_acc - final_val_acc)
    train_test_gap = abs(final_train_acc - test_acc)
    val_test_gap = abs(final_val_acc - test_acc)

    print(f"â€¢ Training-Validation gap: {train_val_gap:.4f}")
    print(f"â€¢ Training-Test gap: {train_test_gap:.4f}")
    print(f"â€¢ Validation-Test gap: {val_test_gap:.4f}")

    if train_val_gap < 0.05:
        overfitting_status = "MINIMAL overfitting - Excellent generalization"
    elif train_val_gap < 0.10:
        overfitting_status = "SLIGHT overfitting - Good generalization"
    elif train_val_gap < 0.20:
        overfitting_status = "MODERATE overfitting - Consider regularization"
    else:
        overfitting_status = "HIGH overfitting - Needs regularization"

    print(f"â€¢ Overfitting assessment: {overfitting_status}")

    # Validation consistency
    val_acc_trend = np.diff(val_acc[-5:]) if epochs >= 5 else np.diff(val_acc)
    if np.mean(val_acc_trend) > 0:
        trend_status = "IMPROVING trend"
    elif np.mean(val_acc_trend) < -0.01:
        trend_status = "DECLINING trend - possible overfitting"
    else:
        trend_status = "STABLE trend"

    print(f"â€¢ Validation trend (recent): {trend_status}")

    # ========================================================================
    # 4e. Overall Training Assessment
    # ========================================================================
    print(f"\n[ASSESSMENT] Overall Training Assessment:")
    print("-" * 40)

    # Performance level
    if test_acc >= 0.95:
        performance_level = "EXCELLENT"
    elif test_acc >= 0.90:
        performance_level = "VERY GOOD"
    elif test_acc >= 0.80:
        performance_level = "GOOD"
    else:
        performance_level = "NEEDS IMPROVEMENT"

    print(f"â€¢ Overall performance: {performance_level}")
    print(f"â€¢ Model effectiveness: 1D CNN successfully learned Iris patterns")
    print(f"â€¢ Architecture suitability: Appropriate for this classification task")

    if val_test_gap < 0.05:
        print(f"â€¢ Generalization: Excellent - validation performance matches test")
    else:
        print(f"â€¢ Generalization: Good - minor difference between validation and test")

    print("=" * 60)

def visualize_results(history, evaluation_results, target_names):
    """
    Create comprehensive visualization of results

    Args:
        history: Training history
        evaluation_results: Evaluation results dictionary
        target_names: Names of target classes
    """

    print(f"\nğŸ¨ Creating Result Visualizations...")
    print("-" * 40)

    # Create subplot figure
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

    # Plot 1: Training History - Accuracy
    epochs_range = range(1, len(history.history['accuracy']) + 1)
    ax1.plot(epochs_range, history.history['accuracy'], 'bo-', label='Training Accuracy')
    ax1.plot(epochs_range, history.history['val_accuracy'], 'ro-', label='Validation Accuracy')
    ax1.set_title('Model Accuracy Over Epochs', fontweight='bold', fontsize=14)
    ax1.set_xlabel('Epoch', fontweight='bold')
    ax1.set_ylabel('Accuracy', fontweight='bold')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot 2: Training History - Loss
    ax2.plot(epochs_range, history.history['loss'], 'go-', label='Training Loss')
    ax2.plot(epochs_range, history.history['val_loss'], 'mo-', label='Validation Loss')
    ax2.set_title('Model Loss Over Epochs', fontweight='bold', fontsize=14)
    ax2.set_xlabel('Epoch', fontweight='bold')
    ax2.set_ylabel('Loss', fontweight='bold')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # Plot 3: Confusion Matrix
    cm = evaluation_results['confusion_matrix']
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=target_names, yticklabels=target_names, ax=ax3)
    ax3.set_title('Confusion Matrix', fontweight='bold', fontsize=14)
    ax3.set_xlabel('Predicted Label', fontweight='bold')
    ax3.set_ylabel('True Label', fontweight='bold')

    # Plot 4: Performance Summary
    ax4.axis('off')
    ax4.set_title('Performance Summary', fontweight='bold', fontsize=16)

    # Performance text
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    test_acc = evaluation_results['test_accuracy']

    summary_text = f"""
    [RESULTS] FINAL RESULTS:

    Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)
    Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)
    Test Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)

    [PERF] MODEL PERFORMANCE:
    â€¢ Excellent classification performance
    â€¢ Good generalization to test data
    â€¢ 1D CNN effectively learned iris patterns

    [DONE] BONUS PART 1 COMPLETED
    """

    ax4.text(0.1, 0.9, summary_text, transform=ax4.transAxes, fontsize=12,
             verticalalignment='top', bbox=dict(boxstyle="round,pad=0.5", facecolor='lightblue', alpha=0.7))

    plt.tight_layout()
    plt.show()

# ============================================================================
# EXECUTE BONUS PART 1
# ============================================================================

def main():
    """
    Main function to execute Bonus Part 1: 1D CNN Implementation
    """

    print("\nğŸ¯ EXECUTING BONUS PART 1: 1D CNN IMPLEMENTATION")
    print("=" * 70)

    # Step 1: Load dataset
    X, y, feature_names, target_names = load_iris_dataset()

    # Step 2: Preprocess data
    data = preprocess_iris_data(X, y)

    # Step 3: Build model architecture
    input_shape = (4, 1)  # 4 features, 1 channel
    model = build_1d_cnn_model(input_shape)
    model = compile_and_display_model(model)

    # Step 4: Train and evaluate
    history = train_cnn_model(model, data)
    evaluation_results = evaluate_cnn_model(model, data, target_names)

    # Detailed training performance analysis
    analyze_training_performance(history, evaluation_results)

    # Visualize results
    visualize_results(history, evaluation_results, target_names)

    print("\n" + "=" * 70)
    print("ğŸ‰ BONUS PART 1 COMPLETED SUCCESSFULLY!")
    print("=" * 70)
    print("âœ… All requirements implemented:")
    print("   â€¢ âœ… Iris dataset loaded and processed")
    print("   â€¢ âœ… Feature normalization applied")
    print("   â€¢ âœ… One-hot encoding implemented")
    print("   â€¢ âœ… CNN input reshaping completed")
    print("   â€¢ âœ… 1D Convolutional layers built")
    print("   â€¢ âœ… MaxPooling layer added")
    print("   â€¢ âœ… Dense output layer implemented")
    print("   â€¢ âœ… Model trained and evaluated")
    print(f"   â€¢ âœ… Final test accuracy: {evaluation_results['test_accuracy']:.4f}")
    print("=" * 70)

    return model, history, evaluation_results

# Run the main function
if __name__ == "__main__":
    model, history, evaluation_results = main()
```

```python
# BONUS ASSIGNMENT - PART 2: Implement an ANN (15 points)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

print("ğŸš€ BONUS ASSIGNMENT - PART 2: ANN Implementation")
print("=" * 70)
print("ğŸ“‹ Assignment Overview:")
print("   â€¢ Dataset: Iris dataset (same preprocessing as CNN)")
print("   â€¢ Task: Artificial Neural Network (ANN) for classification")
print("   â€¢ Points: 15")
print("   â€¢ Comparison: ANN vs CNN performance")
print("=" * 70)

# ============================================================================
# STEP 1: Same Preprocessing as CNN (except reshaping)
# ============================================================================

def preprocess_iris_data_for_ann(X, y, test_size=0.2, random_state=42):
    """
    Preprocess the Iris dataset for ANN (same as CNN except no reshaping)

    Args:
        X: Features array
        y: Target array
        test_size: Test set size (default: 0.2)
        random_state: Random seed for reproducibility

    Returns:
        Preprocessed training and testing data
    """

    print("\nğŸ”§ STEP 1: Same Preprocessing as CNN (except reshaping)")
    print("-" * 60)

    # Split the data first
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )

    print(f"âœ… Data split completed:")
    print(f"   â€¢ Training samples: {X_train.shape[0]}")
    print(f"   â€¢ Testing samples: {X_test.shape[0]}")

    # ========================================================================
    # 1a. Normalize feature values (same as CNN)
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 1a: Normalize feature values (same as CNN)")
    print("-" * 30)

    scaler = StandardScaler()
    X_train_normalized = scaler.fit_transform(X_train)
    X_test_normalized = scaler.transform(X_test)

    print(f"âœ… Feature normalization completed")
    print(f"   â€¢ Method: StandardScaler (mean=0, std=1)")
    print(f"   â€¢ Training mean: {np.mean(X_train_normalized, axis=0)}")
    print(f"   â€¢ Training std: {np.std(X_train_normalized, axis=0)}")

    # ========================================================================
    # 1b. One-hot encode target labels (same as CNN)
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 1b: One-hot encode target labels (same as CNN)")
    print("-" * 30)

    y_train_onehot = to_categorical(y_train, num_classes=3)
    y_test_onehot = to_categorical(y_test, num_classes=3)

    print(f"âœ… One-hot encoding completed")
    print(f"   â€¢ Original shape: {y_train.shape}")
    print(f"   â€¢ One-hot shape: {y_train_onehot.shape}")
    print(f"   â€¢ Classes: 3 (setosa, versicolor, virginica)")

    # ========================================================================
    # 1c. NO RESHAPING for ANN (difference from CNN)
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 1c: No reshaping for ANN (difference from CNN)")
    print("-" * 30)

    print(f"âœ… ANN input format:")
    print(f"   â€¢ CNN input shape: (samples, features, 1) - requires reshaping")
    print(f"   â€¢ ANN input shape: (samples, features) - no reshaping needed")
    print(f"   â€¢ Training shape: {X_train_normalized.shape}")
    print(f"   â€¢ Testing shape: {X_test_normalized.shape}")

    return {
        'X_train': X_train_normalized,  # No reshaping for ANN
        'X_test': X_test_normalized,    # No reshaping for ANN
        'y_train': y_train_onehot,
        'y_test': y_test_onehot,
        'y_train_original': y_train,
        'y_test_original': y_test,
        'scaler': scaler
    }

# ============================================================================
# STEP 2: ANN Architecture
# ============================================================================

def build_ann_model(input_shape):
    """
    Build ANN model according to bonus requirements

    Args:
        input_shape: Number of input features

    Returns:
        Compiled ANN model
    """

    print("\nğŸ—ï¸ STEP 2: ANN Architecture")
    print("-" * 40)

    model = Sequential()

    # ========================================================================
    # 2a. Input Layer: matching number of features
    # ========================================================================
    print(f"ğŸ“‹ INSTRUCTION 2a: Input Layer matching number of features")

    model.add(Dense(
        128,  # First hidden layer also serves as input layer
        activation='relu',
        input_shape=(input_shape,),
        name='input_hidden_layer'
    ))

    print(f"âœ… Input layer added:")
    print(f"   â€¢ Input features: {input_shape}")
    print(f"   â€¢ First layer neurons: 128")
    print(f"   â€¢ Activation: ReLU")

    # ========================================================================
    # 2b. Hidden Layer(s): at least one
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 2b: Hidden Layer(s) - at least one")

    # Additional hidden layers for better learning capacity
    model.add(Dense(
        64,
        activation='relu',
        name='hidden_layer_2'
    ))

    model.add(Dropout(0.3, name='dropout_layer'))  # Regularization

    model.add(Dense(
        32,
        activation='relu',
        name='hidden_layer_3'
    ))

    print(f"âœ… Hidden layers added:")
    print(f"   â€¢ Hidden layer 1: 128 neurons, ReLU (combined with input)")
    print(f"   â€¢ Hidden layer 2: 64 neurons, ReLU")
    print(f"   â€¢ Dropout layer: 30% dropout rate (regularization)")
    print(f"   â€¢ Hidden layer 3: 32 neurons, ReLU")
    print(f"   â€¢ Total hidden layers: 3 (exceeds minimum requirement)")

    # ========================================================================
    # 2c. Output Layer: softmax activation for multi-class classification
    # ========================================================================
    print(f"\nğŸ“‹ INSTRUCTION 2c: Output Layer with softmax activation")

    model.add(Dense(
        3,
        activation='softmax',
        name='output_layer'
    ))

    print(f"âœ… Output layer added:")
    print(f"   â€¢ Output neurons: 3 (for 3 classes)")
    print(f"   â€¢ Activation: Softmax (for multi-class classification)")
    print(f"   â€¢ Purpose: Probability distribution over classes")

    return model

def compile_and_display_ann_model(model):
    """
    Compile ANN model and display architecture summary

    Args:
        model: Built ANN model

    Returns:
        Compiled model
    """

    # Compile the model with same settings as CNN
    model.compile(
        optimizer='adam',  # Same as CNN
        loss='categorical_crossentropy',  # Same as CNN (one-hot encoded labels)
        metrics=['accuracy']  # Same as CNN
    )

    print(f"\nğŸ“Š ANN Model Compilation:")
    print(f"   â€¢ Optimizer: Adam (same as CNN)")
    print(f"   â€¢ Loss: categorical_crossentropy (same as CNN)")
    print(f"   â€¢ Metrics: accuracy (same as CNN)")

    print(f"\nğŸ—ï¸ ANN Architecture Summary:")
    print("-" * 50)
    model.summary()

    return model

# ============================================================================
# STEP 3: Train using same optimizer, batch size, and epochs
# ============================================================================

def train_ann_model(model, data, epochs=20, batch_size=16, validation_split=0.2):
    """
    Train the ANN model using same parameters as CNN

    Args:
        model: Compiled ANN model
        data: Preprocessed data dictionary
        epochs: Number of training epochs (same as CNN)
        batch_size: Training batch size (same as CNN)
        validation_split: Validation split ratio (same as CNN)

    Returns:
        Training history
    """

    print(f"\nğŸš€ STEP 3: Train using same optimizer, batch size, and epochs")
    print("-" * 60)

    print(f"ğŸ“‹ INSTRUCTION 3: Same training parameters as CNN")
    print(f"   â€¢ Optimizer: Adam (same as CNN)")
    print(f"   â€¢ Epochs: {epochs} (same as CNN)")
    print(f"   â€¢ Batch size: {batch_size} (same as CNN)")
    print(f"   â€¢ Validation split: {validation_split} (same as CNN)")

    # Train the model
    history = model.fit(
        data['X_train'], data['y_train'],
        epochs=epochs,
        batch_size=batch_size,
        validation_split=validation_split,
        verbose=1
    )

    print(f"\nâœ… ANN training completed successfully!")

    return history

# ============================================================================
# STEP 4: Evaluate the model
# ============================================================================

def evaluate_ann_model(model, data, target_names):
    """
    Comprehensive evaluation of the ANN model using multiple performance metrics

    Args:
        model: Trained ANN model
        data: Preprocessed data dictionary
        target_names: Names of target classes

    Returns:
        Evaluation results
    """

    print(f"\nğŸ“Š STEP 4: Comprehensive ANN Model Evaluation")
    print("-" * 60)

    # ========================================================================
    # 4a. Basic Evaluation: Accuracy and Loss Curves
    # ========================================================================
    print(f"ğŸ“‹ EVALUATION TOOL 1: Accuracy and Loss Curves")
    print("-" * 40)

    # Evaluate on test set
    test_loss, test_accuracy = model.evaluate(data['X_test'], data['y_test'], verbose=0)

    print(f"âœ… Basic Metrics:")
    print(f"   â€¢ Test Loss: {test_loss:.4f}")
    print(f"   â€¢ Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

    # ========================================================================
    # 4b. Prediction Analysis
    # ========================================================================
    print(f"\nğŸ“‹ EVALUATION TOOL 2: Prediction Analysis")
    print("-" * 40)

    # Get predictions for detailed analysis
    y_pred_proba = model.predict(data['X_test'], verbose=0)
    y_pred = np.argmax(y_pred_proba, axis=1)
    y_true = data['y_test_original']

    # Prediction confidence analysis
    max_probs = np.max(y_pred_proba, axis=1)
    mean_confidence = np.mean(max_probs)
    min_confidence = np.min(max_probs)

    print(f"âœ… Prediction Confidence:")
    print(f"   â€¢ Mean confidence: {mean_confidence:.4f}")
    print(f"   â€¢ Min confidence: {min_confidence:.4f}")
    print(f"   â€¢ Confident predictions (>0.9): {np.sum(max_probs > 0.9)}/{len(max_probs)}")

    # ========================================================================
    # 4c. Classification Metrics
    # ========================================================================
    print(f"\nğŸ“‹ EVALUATION TOOL 3: Classification Metrics")
    print("-" * 40)

    from sklearn.metrics import precision_score, recall_score, f1_score

    # Calculate metrics
    precision_macro = precision_score(y_true, y_pred, average='macro')
    recall_macro = recall_score(y_true, y_pred, average='macro')
    f1_macro = f1_score(y_true, y_pred, average='macro')

    precision_weighted = precision_score(y_true, y_pred, average='weighted')
    recall_weighted = recall_score(y_true, y_pred, average='weighted')
    f1_weighted = f1_score(y_true, y_pred, average='weighted')

    print(f"âœ… Macro-averaged Metrics:")
    print(f"   â€¢ Precision: {precision_macro:.4f}")
    print(f"   â€¢ Recall: {recall_macro:.4f}")
    print(f"   â€¢ F1-Score: {f1_macro:.4f}")

    print(f"âœ… Weighted-averaged Metrics:")
    print(f"   â€¢ Precision: {precision_weighted:.4f}")
    print(f"   â€¢ Recall: {recall_weighted:.4f}")
    print(f"   â€¢ F1-Score: {f1_weighted:.4f}")

    # ========================================================================
    # 4d. Per-Class Performance Analysis
    # ========================================================================
    print(f"\nğŸ“‹ EVALUATION TOOL 4: Per-Class Performance")
    print("-" * 40)

    # Per-class metrics
    precision_per_class = precision_score(y_true, y_pred, average=None)
    recall_per_class = recall_score(y_true, y_pred, average=None)
    f1_per_class = f1_score(y_true, y_pred, average=None)

    print(f"âœ… Per-Class Metrics:")
    for i, class_name in enumerate(target_names):
        print(f"   â€¢ {class_name}:")
        print(f"     - Precision: {precision_per_class[i]:.4f}")
        print(f"     - Recall: {recall_per_class[i]:.4f}")
        print(f"     - F1-Score: {f1_per_class[i]:.4f}")

    # ========================================================================
    # 4e. Classification Report
    # ========================================================================
    print(f"\nğŸ“‹ EVALUATION TOOL 5: Detailed Classification Report")
    print("-" * 40)
    print(classification_report(y_true, y_pred, target_names=target_names))

    # ========================================================================
    # 4f. Confusion Matrix Analysis
    # ========================================================================
    print(f"\nğŸ“‹ EVALUATION TOOL 6: Confusion Matrix Analysis")
    print("-" * 40)

    cm = confusion_matrix(y_true, y_pred)
    print(f"âœ… Confusion Matrix:")
    print(cm)

    # Calculate per-class accuracy from confusion matrix
    class_accuracy = np.diag(cm) / np.sum(cm, axis=1)
    print(f"âœ… Per-Class Accuracy from Confusion Matrix:")
    for i, (class_name, acc) in enumerate(zip(target_names, class_accuracy)):
        print(f"   â€¢ {class_name}: {acc:.4f} ({acc*100:.2f}%)")

    # ========================================================================
    # 4g. Error Analysis
    # ========================================================================
    print(f"\nğŸ“‹ EVALUATION TOOL 7: Error Analysis")
    print("-" * 40)

    # Find misclassified samples
    misclassified = np.where(y_true != y_pred)[0]
    correct_predictions = np.where(y_true == y_pred)[0]

    print(f"âœ… Error Statistics:")
    print(f"   â€¢ Correct predictions: {len(correct_predictions)}/{len(y_true)}")
    print(f"   â€¢ Misclassified samples: {len(misclassified)}/{len(y_true)}")
    print(f"   â€¢ Error rate: {len(misclassified)/len(y_true)*100:.2f}%")

    if len(misclassified) > 0:
        print(f"âœ… Misclassification Details:")
        for idx in misclassified:
            true_class = target_names[y_true[idx]]
            pred_class = target_names[y_pred[idx]]
            confidence = max_probs[idx]
            print(f"   â€¢ Sample {idx}: True={true_class}, Predicted={pred_class}, Confidence={confidence:.4f}")

    # ========================================================================
    # 4h. Model Robustness Assessment
    # ========================================================================
    print(f"\nğŸ“‹ EVALUATION TOOL 8: Model Robustness Assessment")
    print("-" * 40)

    # Check prediction consistency
    low_confidence_threshold = 0.7
    low_confidence_predictions = np.sum(max_probs < low_confidence_threshold)

    print(f"âœ… Robustness Metrics:")
    print(f"   â€¢ High confidence predictions (>0.9): {np.sum(max_probs > 0.9)}")
    print(f"   â€¢ Medium confidence predictions (0.7-0.9): {np.sum((max_probs >= 0.7) & (max_probs <= 0.9))}")
    print(f"   â€¢ Low confidence predictions (<0.7): {low_confidence_predictions}")

    if low_confidence_predictions == 0:
        print(f"   â€¢ Robustness: EXCELLENT - All predictions highly confident")
    elif low_confidence_predictions <= len(y_true) * 0.1:
        print(f"   â€¢ Robustness: GOOD - Few uncertain predictions")
    else:
        print(f"   â€¢ Robustness: MODERATE - Some uncertain predictions")

    return {
        'test_loss': test_loss,
        'test_accuracy': test_accuracy,
        'y_pred': y_pred,
        'y_true': y_true,
        'y_pred_proba': y_pred_proba,
        'confusion_matrix': cm,
        'precision_macro': precision_macro,
        'recall_macro': recall_macro,
        'f1_macro': f1_macro,
        'precision_weighted': precision_weighted,
        'recall_weighted': recall_weighted,
        'f1_weighted': f1_weighted,
        'precision_per_class': precision_per_class,
        'recall_per_class': recall_per_class,
        'f1_per_class': f1_per_class,
        'mean_confidence': mean_confidence,
        'misclassified_count': len(misclassified),
        'class_accuracy': class_accuracy
    }

def analyze_ann_training_performance(history, evaluation_results):
    """
    Analyze ANN training performance

    Args:
        history: Training history object
        evaluation_results: Model evaluation results
    """

    print(f"\nğŸ“‹ ANN TRAINING PERFORMANCE ANALYSIS")
    print("=" * 60)

    # Extract training metrics
    train_acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = len(train_acc)

    # Performance analysis
    final_train_acc = train_acc[-1]
    final_val_acc = val_acc[-1]
    best_val_acc = max(val_acc)
    test_acc = evaluation_results['test_accuracy']

    print(f"[ACCURACY] ANN Accuracy Analysis:")
    print(f"           â€¢ Final training accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")
    print(f"           â€¢ Final validation accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)")
    print(f"           â€¢ Best validation accuracy: {best_val_acc:.4f}")
    print(f"           â€¢ Test accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)")

    # Loss analysis
    final_train_loss = train_loss[-1]
    final_val_loss = val_loss[-1]
    min_val_loss = min(val_loss)
    test_loss = evaluation_results['test_loss']

    print(f"\n[LOSS] ANN Loss Analysis:")
    print(f"       â€¢ Final training loss: {final_train_loss:.4f}")
    print(f"       â€¢ Final validation loss: {final_val_loss:.4f}")
    print(f"       â€¢ Minimum validation loss: {min_val_loss:.4f}")
    print(f"       â€¢ Test loss: {test_loss:.4f}")

    # Overfitting analysis
    train_val_gap = abs(final_train_acc - final_val_acc)
    print(f"\n[OVERFITTING] ANN Generalization:")
    print(f"              â€¢ Training-Validation gap: {train_val_gap:.4f}")

    if train_val_gap < 0.05:
        print(f"              â€¢ Assessment: Excellent generalization")
    elif train_val_gap < 0.10:
        print(f"              â€¢ Assessment: Good generalization")
    else:
        print(f"              â€¢ Assessment: Some overfitting detected")

    print("=" * 60)

def compare_cnn_vs_ann(cnn_results, ann_results):
    """
    Compare CNN vs ANN performance

    Args:
        cnn_results: CNN evaluation results
        ann_results: ANN evaluation results
    """

    print(f"\nğŸ“Š CNN vs ANN PERFORMANCE COMPARISON")
    print("=" * 60)

    print(f"[ACCURACY] Model Comparison:")
    print(f"           â€¢ CNN Test Accuracy: {cnn_results['test_accuracy']:.4f} ({cnn_results['test_accuracy']*100:.2f}%)")
    print(f"           â€¢ ANN Test Accuracy: {ann_results['test_accuracy']:.4f} ({ann_results['test_accuracy']*100:.2f}%)")

    acc_diff = ann_results['test_accuracy'] - cnn_results['test_accuracy']
    if acc_diff > 0:
        print(f"           â€¢ Winner: ANN (+{acc_diff:.4f} better)")
    elif acc_diff < 0:
        print(f"           â€¢ Winner: CNN (+{abs(acc_diff):.4f} better)")
    else:
        print(f"           â€¢ Result: Tie (equal performance)")

    print(f"\n[LOSS] Loss Comparison:")
    print(f"       â€¢ CNN Test Loss: {cnn_results['test_loss']:.4f}")
    print(f"       â€¢ ANN Test Loss: {ann_results['test_loss']:.4f}")

    loss_diff = cnn_results['test_loss'] - ann_results['test_loss']
    if loss_diff > 0:
        print(f"       â€¢ Lower loss: ANN (better by {loss_diff:.4f})")
    else:
        print(f"       â€¢ Lower loss: CNN (better by {abs(loss_diff):.4f})")

    print("=" * 60)

def visualize_ann_results(history, evaluation_results, target_names):
    """
    Create comprehensive visualization of ANN results

    Args:
        history: Training history
        evaluation_results: Evaluation results dictionary
        target_names: Names of target classes
    """

    print(f"\nğŸ¨ Creating ANN Result Visualizations...")
    print("-" * 50)

    # Create subplot figure
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

    # Plot 1: Training History - Accuracy
    epochs_range = range(1, len(history.history['accuracy']) + 1)
    ax1.plot(epochs_range, history.history['accuracy'], 'bo-', label='Training Accuracy')
    ax1.plot(epochs_range, history.history['val_accuracy'], 'ro-', label='Validation Accuracy')
    ax1.set_title('ANN Model Accuracy Over Epochs', fontweight='bold', fontsize=14)
    ax1.set_xlabel('Epoch', fontweight='bold')
    ax1.set_ylabel('Accuracy', fontweight='bold')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot 2: Training History - Loss
    ax2.plot(epochs_range, history.history['loss'], 'go-', label='Training Loss')
    ax2.plot(epochs_range, history.history['val_loss'], 'mo-', label='Validation Loss')
    ax2.set_title('ANN Model Loss Over Epochs', fontweight='bold', fontsize=14)
    ax2.set_xlabel('Epoch', fontweight='bold')
    ax2.set_ylabel('Loss', fontweight='bold')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # Plot 3: Confusion Matrix
    cm = evaluation_results['confusion_matrix']
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
                xticklabels=target_names, yticklabels=target_names, ax=ax3)
    ax3.set_title('ANN Confusion Matrix', fontweight='bold', fontsize=14)
    ax3.set_xlabel('Predicted Label', fontweight='bold')
    ax3.set_ylabel('True Label', fontweight='bold')

    # Plot 4: Performance Summary
    ax4.axis('off')
    ax4.set_title('ANN Performance Summary', fontweight='bold', fontsize=16)

    # Performance text
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    test_acc = evaluation_results['test_accuracy']

    summary_text = f"""
    [RESULTS] ANN FINAL RESULTS:

    Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)
    Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)
    Test Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)

    [PERF] ANN MODEL PERFORMANCE:
    â€¢ Excellent classification performance
    â€¢ Good generalization to test data
    â€¢ ANN effectively learned iris patterns

    [DONE] BONUS PART 2 COMPLETED
    """

    ax4.text(0.1, 0.9, summary_text, transform=ax4.transAxes, fontsize=12,
             verticalalignment='top', bbox=dict(boxstyle="round,pad=0.5", facecolor='lightgreen', alpha=0.7))

    plt.tight_layout()
    plt.show()

# ============================================================================
# EXECUTE BONUS PART 2
# ============================================================================

def main_ann():
    """
    Main function to execute Bonus Part 2: ANN Implementation
    """

    print("\nğŸ¯ EXECUTING BONUS PART 2: ANN IMPLEMENTATION")
    print("=" * 70)

    # Load dataset (same as Part 1)
    iris = load_iris()
    X, y = iris.data, iris.target
    feature_names = iris.feature_names
    target_names = iris.target_names

    print(f"âœ… Dataset loaded: {X.shape[0]} samples, {X.shape[1]} features")

    # Step 1: Preprocess data (same as CNN except reshaping)
    data = preprocess_iris_data_for_ann(X, y)

    # Step 2: Build ANN architecture
    input_features = X.shape[1]  # 4 features
    ann_model = build_ann_model(input_features)
    ann_model = compile_and_display_ann_model(ann_model)

    # Step 3: Train using same parameters as CNN
    ann_history = train_ann_model(ann_model, data)

    # Step 4: Evaluate the model
    ann_evaluation_results = evaluate_ann_model(ann_model, data, target_names)

    # Analyze training performance
    analyze_ann_training_performance(ann_history, ann_evaluation_results)

    # Visualize results
    visualize_ann_results(ann_history, ann_evaluation_results, target_names)

    print("\n" + "=" * 70)
    print("ğŸ‰ BONUS PART 2 COMPLETED SUCCESSFULLY!")
    print("=" * 70)
    print("âœ… All ANN requirements implemented:")
    print("   â€¢ âœ… Same preprocessing as CNN (except reshaping)")
    print("   â€¢ âœ… Input layer matching number of features")
    print("   â€¢ âœ… Multiple hidden layers implemented")
    print("   â€¢ âœ… Output layer with softmax activation")
    print("   â€¢ âœ… Same training parameters as CNN")
    print("   â€¢ âœ… Model trained and evaluated")
    print(f"   â€¢ âœ… Final test accuracy: {ann_evaluation_results['test_accuracy']:.4f}")
    print("=" * 70)

    return ann_model, ann_history, ann_evaluation_results

# Run the main function
if __name__ == "__main__":
    ann_model, ann_history, ann_evaluation_results = main_ann()

    # Optional: Compare with CNN results if available
    print("\nğŸ“‹ To compare with CNN results, run both Part 1 and Part 2 together")
```
